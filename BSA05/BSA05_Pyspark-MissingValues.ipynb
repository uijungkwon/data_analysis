#### Pyspark handling missing values
- Dropping columns
- Dropping rows
- Various parameter in dropping functionalities
- Handling missing values by mean, median, and mode

```python
from pyspark.sql import SparkSession 
스파크 = SparkSession.builder.appName('Missing').getOrCreate()
스파크.conf.set("spark.sql.execution.arrow.enabled", "true")#메모리 문제 해결
```

```python
스파크DF = 스파크.read.csv("hdfs://localhost:9000/Spark/Employee_missing.csv", header=True, encoding="cp949", inferSchema=True) # 스파크 폴더안에 있는 데이터를 가져옴 
스파크DF.show()
```

```python
### drop the columns
스파크DF.drop('id').show() ##id 칼럼을 제거 
```

#### 결측값처리
- 행 삭제
  - DataFrame.dropna([how,thresh,subset])
  - DataFrame.na.drop()
  - default:how=any
- 대체
  - DataFrame.fillna(Value,[,subset])
  - DataFrame.na.fill()
  - pyspark.ml.feature.Imputer()
 

```python
### drop the rows
스파크DF.na.drop().show()
스파크DF.dropna().show()
```

```python
스파크DF.na.drop(how='all').show()
스파크DF.dropna(how='all').show()
```

```python
### threshold - 예를들어 결측값이 아닌 데이터 개수가 2 미만이면 제거
스파크DF.na.drop(thresh=4).show()
스파크DF.dropna(thresh=2).show()
```

```python
### subset
스파크DF.na.drop(subset=['jobtime']).show()
스파크DF.dropna(subset=['jobtime']).show()
```

```python
### Filling the missing values-> 특정값 대체 
스파크DF.na.fill('NA').show() ## 데이터의 값 타입이 문자 인 것만 'NA' 로 변경됨 -> 변경할 값의 타입이 문자이므로!
스파크DF.fillna('NA').show()
```

```python
스파크DF.na.fill('NA', 'jobtime').show()
스파크DF.fillna('NA', 'jobtime').show() ## jobtime은 숫자 타입이므로 문자열인 'NA'로 대체 되지 않음!!
```

```python
# (연구실 gender만, 집 gender, jobtime)
스파크DF.na.fill('NA', ['gender','jobtime']).show()
스파크DF.fillna('NA',['gender','jobtime']).show()
```

#### 결측값을 통계값으로 대체

```python
### 수치자료 대체
from pyspark.ml.feature import Imputer # 대체하는 imputer 값을 가져옴 
```

```python
대체 = Imputer(
     inputCols=['educ', 'salary'], # inputCols: 대체할 칼럼 -> 각 칼럼의 평균으로 결측값을 대체할 것임 
     outputCols=["{}_imputed".format(c) for c in ['educ', 'salary']]#시험에 나오진 않음 
     ).setStrategy("mean")## 두개의 칼럼을 새로 만들고 결측값을 "평균"값으로 대체함 
```

```python
### Add imputation columns to dataframe
대체.fit(스파크DF).transform(스파크DF).show()## fit():평균을 계산하는 과정, transform():계산한 값으로 대체하는 과정 !(데이터가 반드시 같을 필요는 없음)
```

```python
대체 = Imputer(
     inputCols=['educ', 'salary'],
     outputCols=["{}_imputed".format(c) for c in ['educ', 'salary']]
     ).setStrategy("mode")##mode:최빈값을 계산함
대체.fit(스파크DF).transform(스파크DF).show()
```

```python
스파크DF.dtypes ## 각 칼럼의 타입값을 나열함 -> 숫자인 데이터로 결측값 대체하기 위해서 
```

```python
from pyspark.sql.types import StringType
수치변수 = [변수.name for 변수 in 스파크DF.schema.fields if not isinstance(변수.dataType, StringType)]
수치변수
##for 변수 in 스파크DF.schema.fields  : 스파크DF에서 칼럼들을 가져와서 "변수" 값에 대입함
## 저장한 변수의 타입이 문자열이 아니라면, 해당 변수의 name을 "수치변수" 값에 대입함
##
```

```python
스파크DF[수치변수].show() ## 수치변수에 대해서만 보여주고 있음 
```

```python
대체 = Imputer(inputCols=수치변수, outputCols=수치변수).setStrategy("mean")
대체.fit(스파크DF).transform(스파크DF).show()## 원래의 스파크DF에 수치변수를 하나씩 대입하여 계산한 후, 변환하여 보여주고있음
```
