## matplotlib에서 한글처리

````python
import matplotlib.pyplot as plt ##그림 그릴 수 있는 모듈을 가져옴
import matplotlib 
````

````python
plt.plot([1,2,3],[4,5,8]) ## 선그래프 (x,y)
plt.xlabel("문자")## x축에 한글 이름을 지정함
````

## 간단한 한글처리방법

### 1. C:\Windows\fonts에 있는 한글 폰트 사용
- matplotlib의 rc 모듈 사용

```Python
import matplotlib
matplotlib.rc('font',family='NanumGothic')##NanumGothic에는 한글이 존재 ### Mac에서는 AppleGothic
```

### 2. matplotlib에서 관리하는 폰트 폴드에 있는 한글 폰트사용
- print(matplotlib.matplotlib_fname())로 위치 확인 ## C:\Users\SM-PC\AppData\Local\Programs\Python\Python39\Lib\site-packages\matplotlib\mpl-data\fonts\ttf
- matplotlib의 rcParams 사용

```Python
import matplotlib
matplotlib.rcParams["font.family"] = 'NanumGothic'
```

````python
print(matplotlib.matplotlib_fname())
````

````python
matplotlib.rcParams["font.family"] = 'NanumGothic' ## 나눔 고딕을 설치함, 한글 사용할 수 있음 -> 해결방안 1
plt.plot([1,2,3],[4,5,8])
plt.xlabel("문자")
````

````python
matplotlib.rc('font',family ='NanumGothic')
plt.plot([1,2,3],[4,5,8])
plt.xlabel("문자")
````
