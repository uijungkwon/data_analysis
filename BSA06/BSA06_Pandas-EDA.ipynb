## 그룹별 분석

```python
# 그룹별 평균
팁.groupby('sex').mean() # 성별을 기준으로 칼럼을 나누어 평균을 구함 - 수치데이터에 대해서 
```

### df를 groupby()를 통해 부분 df로 나눈 후 함수적용

```python
팁.groupby('sex').head() # 팁.groupby('sex')을 했을 때, 성별은 두 개의 범주로 나누어져있으니 두개의 데이터프레임을 만듦 
```

```python
for 성별, 그룹 in 팁.groupby('sex'): # 데이터프레임에 해당하는 것을 "그룹"에 대입함 
    print(성별)
    display(그룹.head()) ## 각 성별로 나누어진 그룹을 의미함, display 함수에 넣어야지만 사용자에게 보여짐 
```

### 적용가능한 통계함수
- count : 표본크기
- sum : 합
- mean : 평균
- median : 중앙값
- var, std : 분산, 표준편차
- min, max :  최솟값, 최댓값
- unique, nunique : 고유값, 고유값개수
- prod : 곱
- first, last : 첫번째 값, 마지막 값

### .agg()를 활용하여 다중 통계 표시가능

```python
팁.groupby('sex').agg(['mean', 'std']) # 두개 이상의 함수를 적용할 경우 .agg() 는 묶어주는 역할을 함 
```

```python
팁.groupby(['sex','time']).agg(['mean', 'std'])
```

### 컬럼별로 다룬 통계를 계산할 수 있음

```python
팁.groupby('sex').agg({'total_bill': ['mean','std'], 'tip': 'median', 'size': 'sum' }) ## 적용하는 함수의 값을 변수별로 다르게 하고 싶은 경우(딕셔너리 형태로 지정)
```

```python
# 소수점 자리 수 지정
팁.groupby('sex').agg({'total_bill': ['mean','std'], 'tip': 'median', 'size': 'sum' }).round(2) ## 소수점 둘째자리까지 표시
```

```python
###  "사용자 정의 함수"는 .agg()에 적용
```

```python
# CV(변동계수)
팁.groupby('sex')[['total_bill', 'tip']].agg(lambda x: x.std()/x.mean()) ##CV: 표준편차 / 평균, x: 잘라진 데이터프레임이 통채로 들어옴
```

```python
타이타닉 = sns.load_dataset('titanic') # 타이타닉 데이터 자료를 가져옴 
타이타닉.head(20)
```

```python
타이타닉.isna().sum() ## 결측값 확인 
```

### 그룹별 평균으로 결측값 대체

```python
타이타닉.groupby(['sex','class'],group_keys=False)['age'].apply(lambda x: x.fillna(x.mean()))
## 성별과 등급을 기준으로 age값을 나누어 데이터프레임을 생성한 후 , 데이터 프레임 별로 평균을 계산하여 결측치에 대체하는 것을 apply
```

```python
## 그룹별로 표준화 -> 소프로젝트 주제로 결정할 수 있음 
팁.groupby(['sex','time'],group_keys=False)['tip'].apply(lambda x: (x-x.mean())/x.std()) ## 표준화 공식
#groupby로 데이터프레임을 나눈 후 각 데이터프레임의 표준화 값을 구하여 apply함!
```
