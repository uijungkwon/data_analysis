```python
import requests
import json #데이터의 형태
```

```python
url = "https://api.coinbase.com/v2/exchange-rates?currency=BTC"
```

```python
#Bitcoin의 가격
response = requests.get(url)
비트코인 = json.loads(response.text)#json 파일을 로드함
비트코인
```

```python
print("상품수", len(비트코인['data']['rates']))
print("비트코인가격:", 비트코인['data']['rates']['KRW'])
```

```python
url = "https://api.coinbase.com/v2/exchange-rates?currency=KRW" ##BTC->KRW로 변경
response = requests.get(url)
원화 = json.loads(response.text)['data']['rates']#1원당 해당상품의 가격
원화
```

```python
관심상품 = ['USD', 'EUR', 'JPY', 'CNY', 'BTC', 'ETH', 'DOGE']
```

```python
가격 = []
for 상품 in 관심상품:
    가격 += [1.0/float(원화[상품])] #"가격"리스트 값을 추가함
print(가격)
```

```python
#함수 작성 - 입력된 url에서 자료를 읽고 해당 상품에 연계된 관심상품의 가격 출력
import time #시간관련된 패키지

def 가격산출(url, 관심상품):
    response = requests.get(url)
    환전자료 = json.loads(response.text)['data']['rates']
    가격 = [str(time.asctime())] # 현재시간을 문자열로 표시
    for 상품 in 관심상품:
        가격 += [1.0/float(환전자료[상품])]
    return 가격
```

```python
결과 = 가격산출(url, 관심상품)
print(결과)
```

```python
#지정된 시간에 작업을 수행
from apscheduler.schedulers.background import BackgroundScheduler
```

```python
sched = BackgroundScheduler
def 가격산출():
    url = "https://api.coinbase.com/v2/exchange-rates?currency=KRW"
    관심상품 = ['USD', 'EUR', 'JPY', 'CNY', 'BTC', 'ETH', 'DOGE'] ##데이터를 보여주는 것이 아니라 저장함
    저장파일 = open("원화-환율.csv", "a", encoding = "utf-8")##기존 데이터에 추가 : "a"
    response = requests.get(url)
    환전자료 = json.loads(response.text)['data']['rates']
    가격 = str(time.asctime())
    for 상품 in 관심상품:
        가격 += "," + str(1.0/float(환전자료[상품]))
    print(가격)
    
    저장파일.write(가격+'\n')
    저장파일.close()#반드시 파일 닫기
```

```python
sched.start()
```

```python
from apscheduler.triggers.interval import IntervalTrigger
```

```python

sched.add_job(가격산출, IntervalTrigger(), seconds = 60)##60초 단위로 "가격 산출"작업을 수행
```

```python
##작업중단
sched.shutdown(가격산출)
```
